# Netflix Movies & TV Shows - Data Visualization

### CodeAlpha Data Analytics Internship

**Project:** Data Visualization

**Dataset:** Netflix Movies and TV Shows

**Tools Used:** Python, Pandas, Matplotlib, Seaborn

**Objective:**
Create meaningful visualizations to understand Netflix's content distribution and trends.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (10,6)
```

```python
df = pd.read_csv("../dataset/netflix_titles_cleaned.csv")

df.head()
```

```python
plt.figure(figsize=(7,5))

sns.countplot(x="type", data=df, palette="Set2")

plt.title("Movies vs TV Shows on Netflix", fontsize=16)
plt.xlabel("Content Type")
plt.ylabel("Count")

plt.tight_layout()
plt.savefig("../images/movies_vs_tvshows.png", dpi=300)

plt.show()
```

```python
top_countries = df["country"].value_counts().head(10)

plt.figure(figsize=(10,6))

sns.barplot(
    x=top_countries.values,
    y=top_countries.index,
    palette="viridis"
)

plt.title("Top 10 Countries Producing Netflix Content", fontsize=16)
plt.xlabel("Number of Titles")
plt.ylabel("Country")

plt.tight_layout()
plt.savefig("../images/top10_countries.png", dpi=300)

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.countplot(
    y="rating",
    data=df,
    order=df["rating"].value_counts().index,
    palette="magma"
)

plt.title("Distribution of Content Ratings", fontsize=16)
plt.xlabel("Count")
plt.ylabel("Rating")

plt.tight_layout()
plt.savefig("../images/content_ratings.png", dpi=300)

plt.show()
```

```python
top_genres = df["listed_in"].value_counts().head(10)

plt.figure(figsize=(12,6))

sns.barplot(
    x=top_genres.values,
    y=top_genres.index,
    palette="coolwarm"
)

plt.title("Top 10 Genres on Netflix", fontsize=16)
plt.xlabel("Number of Titles")
plt.ylabel("Genre")

plt.tight_layout()
plt.savefig("../images/top10_genres.png", dpi=300)

plt.show()
```

```python
df["date_added"] = pd.to_datetime(df["date_added"], errors="coerce")

content_year = df["date_added"].dt.year.value_counts().sort_index()

plt.figure(figsize=(12,6))

plt.plot(content_year.index, content_year.values, marker="o", linewidth=2)

plt.title("Netflix Content Added Over the Years", fontsize=16)
plt.xlabel("Year")
plt.ylabel("Number of Titles")

plt.tight_layout()
plt.savefig("../images/content_added_by_year.png", dpi=300)

plt.show()
```

```python
top_directors = df["director"].value_counts().head(10)

plt.figure(figsize=(12,6))

sns.barplot(
    x=top_directors.values,
    y=top_directors.index,
    palette="crest"
)

plt.title("Top 10 Directors on Netflix", fontsize=16)
plt.xlabel("Number of Titles")
plt.ylabel("Director")

plt.tight_layout()
plt.savefig("../images/top10_directors.png", dpi=300)

plt.show()
```

```python
release_year = df["release_year"].value_counts().sort_index()

plt.figure(figsize=(12,6))

plt.plot(release_year.index, release_year.values, color="green", linewidth=2)

plt.title("Netflix Content Release Trend")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")

plt.tight_layout()
plt.savefig("../images/release_year_trend.png", dpi=300)

plt.show()
```

```python
type_count = df["type"].value_counts()

plt.figure(figsize=(7,7))

plt.pie(
    type_count,
    labels=type_count.index,
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Movies vs TV Shows")

plt.tight_layout()
plt.savefig("../images/movies_vs_tv_pie.png", dpi=300)

plt.show()
```
